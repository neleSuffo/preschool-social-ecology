# RQ 2. How much language does the key child produce (utterances and words)?

```python
import sys
from pathlib import Path

# Get the src directory (2 levels up from current notebook location)
src_path = Path(__file__).parent.parent.parent if '__file__' in globals() else Path.cwd().parent.parent
sys.path.append(str(src_path))

from constants import ResearchQuestions
from rq_02_queries import map_vocalizations_to_segments
```

## Map Vocalizations to Social Segments

```python
utterance_data = map_vocalizations_to_segments()
utterance_data.to_csv(ResearchQuestions.UTTERANCE_SEGMENTS_CSV, index=False)
```

## Create Aggregates by Child, Age at recording and Interaction Type

```python
# Aggregate words by child, interaction type, and age at recording and sort the result
word_summary = utterance_data.groupby(['child_id', 'age_at_recording', 'interaction_type'])['words'].sum().reset_index()
word_summary = word_summary.sort_values(['child_id', 'age_at_recording', 'interaction_type'])

word_summary.to_csv(ResearchQuestions.WORD_SUMMARY_CSV, index=False)
```
